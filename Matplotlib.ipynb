```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1,101)       # array in range(1,101)
y = np.random.randint(1,10,100)   # Write random no.in between range(1-10)
```

```python
fig = plt.figure(figsize=(10,3))   # we can adjust figure size (length, height)
plt.plot (x,y, color='green')  # Line Chart
plt.xlim(0,100)
plt.ylim(1,10)
plt.xlabel('X Axis', color='red', size=20)
plt.ylabel("Y Axis", color = 'red', size=20)
plt.title('Line Plot', color='orange')        # Automatically value of size is taken as 10
plt.suptitle('Sales Comparison', size = 20, y = 1.05, color ='blue')   # Size is font of suptitle, y is spacing between the sales comparison and line plot

plt.savefig('Line_plot')
plt.show()
```

```python
fig.savefig('Line_Plot.png', dpi= 300, bbox_inches='tight')     # dpi(Dots per inches) determine how many pixel fig comprises, bbox_inches is removed all the extra white space around the figure
```

```python
#plt.figure(figsize(30,30))
fig, ax = plt.subplots(nrows=2, ncols=2,figsize = (10,10))  # Comand for multiple plots 

ax[0,0].plot(x,y,color='green', linestyle='dashed')
ax[0,0].set_title('Line')
ax[1,0].set_title('Scatter')
ax[1,0].scatter(x,y)
ax[1,1].hist(y,color='orange')
ax[1,1].set_title('Histogram')
ax[0,1].set_title('Pie Chart')
ax[0,1].pie(y)
plt.show()


```
